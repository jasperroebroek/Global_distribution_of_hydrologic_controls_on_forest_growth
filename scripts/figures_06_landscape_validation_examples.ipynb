```python
import thesis
import geomappy as mp
import matplotlib.pyplot as plt
```

```python
loc = "data/landscape_classes/landscape_classes.tif"
cmap = ["White", '#004dac', '#729116', '#b1bc1d', '#e7de23', '#af9a15', '#785707', '#fff9f2']
labels = labels = ["NaN", 'Wetland and open water', 'Lowland', 'Undulating', 'Hilly', 'Low mountainous', 'Mountainous',
                   'High mountainous']
M_landscape = mp.Raster(loc)
```

```python
M_landscape.plot_classified_map([3, 50, 7, 54], basemap=True, legend='colorbar', labels=labels,
                            basemap_kwargs={'resolution': "10m", 'xticks': 2, 'yticks': 2},
                            bins=[0, 1, 2, 3, 4, 5, 6, 7], colors=cmap)
plt.tight_layout()
plt.subplots_adjust(right=0.8)
plt.savefig("figures/si_landscape_example_NL.png", dpi=300)
plt.show()
```
